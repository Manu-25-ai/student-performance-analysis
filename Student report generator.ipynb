```python
# Real-World Project: Student Report Generator – Read CSV of student marks, calculate grades, and save new CSV.
```

```python
import pandas as pd
import numpy as np
```

```python
# Reading CSV
Marks=pd.read_csv("marksheet.csv")
Marks.head()
```

```python
# Calculating Average marks of students 
Marks["Avg_scores"]=(Marks["Science"]+Marks["English"]+Marks["History"]+Marks["Maths"])/4
Marks
```

```python
# Assigning grades based on average marks of students
def grade(x):
    if x >= 60 and x < 100:
        return "A"
    elif x >= 40 and x < 60:
        return "B"
    else:
        return "C"

Marks["Grades"] = Marks["Avg_scores"].apply(grade)
Marks.head()  
```

```python
# Saving modified CSV file
Marks.to_csv("updated_students_data.csv", index=False)
```
